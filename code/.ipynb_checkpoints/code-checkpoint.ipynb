## Background

According to the [Meteorological Services Singapore](http://www.weather.gov.sg/climate-climate-of-singapore/#:~:text=Singapore%20is%20situated%20near%20the,month%2Dto%2Dmonth%20variation.), Singapore has typical tropical climate with adundant rainfall, high and uniform temperatures and high humidity all year round, since its situated near the equator. There are many factors that help us understand the climate of a country and in this project we are going to look into a few, especially rainfall.

Singapore’s climate is characterised by two main monsoon seasons separated by inter-monsoonal periods.  The **Northeast Monsoon** occurs from December to early March, and the **Southwest Monsoon** from June to September.

The major weather systems affecting Singapore that can lead to heavy rainfall are:

- Monsoon surges, or strong wind episodes in the Northeast Monsoon flow bringing about major rainfall events;

- Sumatra squalls, an organised line of thunderstorms travelling eastward across Singapore, having developed over the island of Sumatra or Straits of Malacca west of us;

- Afternoon and evening thunderstorms caused by strong surface heating and by the sea breeze circulation that develops in the afternoon.

Singapore’s climate station has been located at several different sites in the past 140 years. The station had been decommissioned at various points in the past due to changes to local land use in the site’s vicinity, and had to be relocated. Since 1984, the climate station has been located at **Changi**.

There are other metrics of climate such as temperature, humidity, sun shine duration, wind speed, cloud cover etc. All the dataset used in the project comes from [data.gov.sg](data.gov.sg), as recorded at the Changi climate station.

### Coding Challenges

1. Manually calculate mean:

    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!
    
    *Note*: Do not use any mean methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Given a list, calculate mean
def cal_mean(number_list):
    sum = 0
    no_data = 0
    
    for num in number_list:
        sum += num
        no_data += 1
    return sum/no_data

test_list = [1, 2, 3, 4, 5]
result = cal_mean(test_list)
result
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
#Given a list, calculate standard deviation

def cal_sd(number_list):
    
    mean_square_difference = 0
    no_data = 0
    mean = cal_mean(number_list)
    
    for num in number_list:
        mean_square_difference += (num - mean) ** 2
        no_data += 1
    
    variance = mean_square_difference/no_data
    sd = variance ** 0.5
    return sd

test_list = [1, 2, 3, 4, 5, 6]
result = cal_sd(test_list)
result
```

#### Q3.1 Which month have the highest and lowest total rainfall in 1990, 2000, 2010 and 2020?

```python
#Convert values in 'month' to DateTime Format
df_combined['month'] = pd.to_datetime(df_combined['month'])

def find_extreme_rainfall(df, years):
    results_max = []
    results_min = []
    
    for year in years:
        rainfall_data = df[df['month'].dt.year == year]
        max_rainfall = rainfall_data['maximum_rainfall_in_a_day'].max()
        min_rainfall = rainfall_data['maximum_rainfall_in_a_day'].min()
        results_max.append(max_rainfall)
        results_min.append(min_rainfall)
    
    return results_max, results_min

# Specify the years
years = [1990, 2000, 2010, 2020]

# Call the function to obtain maximum and minimum rainfall values
max_rainfall, min_rainfall = find_extreme_rainfall(df_combined, years)

# Print the results
for i, year in enumerate(years):
    print(f"Highest rainfall in {year}: {max_rainfall[i]}")
    print(f"Lowest rainfall in {year}: {min_rainfall[i]}")
    print("")
```

#### Q3.2 Which year have the highest and lowest total rainfall in the date range of analysis?


```python
# Get the unique years from the 'month' column
unique_years = df_combined['month'].dt.year.unique()

# Initialize an empty dictionary to store the yearly totals
yearly_totals = {}

# Iterate over the unique years
for year in unique_years:
    # Filter the dataframe for rows belonging to the current year
    yearly_data = df_combined[df_combined['month'].dt.year == year]
    
    # Calculate the sum of 'total_rainfall' for the current year
    yearly_total = yearly_data['total_rainfall'].sum()
    
    # Store the yearly total in the dictionary
    yearly_totals[year] = yearly_total

    
# Find the highest and lowest yearly totals
max_year = max(yearly_totals, key=yearly_totals.get)
min_year = min(yearly_totals, key=yearly_totals.get)

# Print the highest and lowest yearly totals
print(f"Highest Total Rainfall: Year {max_year}, Total = {yearly_totals[max_year]}")
print(f"Lowest Total Rainfall: Year {min_year}, Total = {yearly_totals[min_year]}")
```

#### 3.3 Which month have the highest and lowest number of rainy days in 1990, 2000, 2010 and 2020?

```python
#Convert values in 'month' to DateTime Format
df_combined['month'] = pd.to_datetime(df_combined['month'])

def find_extreme_no_rainy_days(df, years):
    results_2_max = []
    results_2_min = []
    
    for year in years:
        rainfall_data = df[df['month'].dt.year == year]
        max_no_rainy_days = rainfall_data['no_of_rainy_days'].max()
        min_no_rainy_days = rainfall_data['no_of_rainy_days'].min()
        results_2_max.append(max_no_rainy_days)
        results_2_min.append(min_no_rainy_days)
    
    return results_2_max, results_2_min

# Specify the years
years = [1990, 2000, 2010, 2020]

# Call the function to obtain maximum and minimum rainfall values
max_no_rainy_days, min_no_rainy_days = find_extreme_no_rainy_days(df_combined, years)

# Print the results
for i, year in enumerate(years):
    print(f"Highest rainfall in year {year}: {max_no_rainy_days[i]}")
    print(f"Lowest rainfall in year {year}: {min_no_rainy_days[i]}")
    print("")
```

#### Q3.4 Which year have the highest and lowest number of rainy days in the date range of analysis?

```python
# Initialize an empty dictionary to store the yearly totals
yearly_totals_rainy_days = {}

# Iterate over the unique years
for year in unique_years:
    # Filter the dataframe for rows belonging to the current year
    yearly_data_rainy_days = df_combined[df_combined['month'].dt.year == year]
    
    # Calculate the sum of 'no_of_rainy_days' for the current year
    yearly_total_rainy_day = yearly_data_rainy_days['no_of_rainy_days'].sum()
    
    # Store the yearly total in the dictionary
    yearly_totals_rainy_days[year] = yearly_total_rainy_day

# Find the highest and lowest yearly totals
max_year = max(yearly_totals_rainy_days, key=yearly_totals_rainy_days.get)
min_year = min(yearly_totals_rainy_days, key=yearly_totals_rainy_days.get)

# Print the highest and lowest yearly totals
print(f"Highest Number of Rainy Days: Year {max_year}, Total = {yearly_totals_rainy_days[max_year]}")
print(f"Lowest Number of Rainy Days: Year {min_year}, Total = {yearly_totals_rainy_days[min_year]}")
```

#### Q3.5 Are there any outliers months in the dataset?

```python
def detect_outliers(df, column):
    # Calculate the first quartile (Q1) and third quartile (Q3)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1
    
    # Calculate the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Detect outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    return outliers
```

```python
# Detect outliers in 'maximum_rainfall_in_a_day' column
outliers = detect_outliers(df_combined, 'maximum_rainfall_in_a_day')

# Print the outliers
print(outliers)
```

#### Q2. Visualize distributions using histograms. If you have a lot, consider writing a custom function and use subplots.

```python
def draw_histograms(df):
    # Get the column names
    columns = df.columns
    
    # Determine the number of rows and columns for subplots
    n_rows = (len(columns) + 1) // 2
    n_cols = 2
    
    # Create a figure and subplots
    fig, axs = plt.subplots(n_rows, n_cols, figsize=(10, 10))
    
    # Flatten the subplots array if there is only one row
    if n_rows == 1:
        axs = axs.flatten()
    else:
        axs = axs.reshape(-1)
    
    # Iterate over the columns and draw histograms
    for i, column in enumerate(columns):
        ax = axs[i]
        ax.hist(df[column], bins=25)
        ax.set_title(column)
        ax.set_xlabel('Value')
        ax.set_ylabel('Frequency')
        ax.tick_params(axis='x', rotation=90)
    
    # Adjust the layout
    plt.tight_layout()
    
    # Show the plot
    plt.show()

# Call the function with your DataFrame
draw_histograms(df_combined)
```

#### (1) 'Month' column is uniformly distributed.

#### The remaining 5 columns are approximatelly normally distributed. In particular,
#### (2) Maximum rainfall and total rainfall are right-skewed and humidity is slightly left-skewed.

```python
# Define a function to plot scatter plots
def plot_scatter(df, x, y):
    # Set the style of the plot
    sns.set(style="whitegrid")
    
    # Create a scatter plot
    sns.scatterplot(data=df, x=x, y=y)
    
    # Set the title and labels of the plot
    plt.title(f"Scatter Plot: {x} vs {y}")
    plt.xlabel(x)
    plt.ylabel(y)
    
    # Show the plot
    plt.show()
```

```python
# Plot scatter plots for each pair of features
features = df_combined.columns[1:]  # Exclude the 'month' column
for i in range(len(features)):
    for j in range(i + 1, len(features)):
        x = features[i]
        y = features[j]
        plot_scatter(df_combined, x, y)
```

#### Conclusions

- (1) November to January are the wettest season in Singapore, supported by number of rainy days, total rainfall and relative humidity. December is the wettest month of Singapore by the same 3 measures. 

- (2) Singapore does not get wetter when she is going through the Southwest Monsoon Season (June to September), both in terms of number of rainy days and total rainfall.

- (3) Hottest month is May. This is likely due to May typically experiences high solar insolation, which refers to the amount of solar radiation received on the Earth's surface.
